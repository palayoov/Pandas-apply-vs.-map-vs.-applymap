```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
jobs = pd.read_csv('NYC_Jobs.csv')
```

```python
jobs.head()
```

```python
# Convert any columns that have nurse in the Business Title	to just Nurse

jobs['Business Title'].apply(lambda x: ('Nurse' if 'Nurse' in x else x))
```

```python
def apply_everywhere(row):
    return str(row) + ' ' + "It's 2024!"

jobs.applymap(func=apply_everywhere)
```

```python
jobs.groupby('Agency')['# Of Positions'].count()
```
